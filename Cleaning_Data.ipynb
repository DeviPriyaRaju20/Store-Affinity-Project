```python
import pandas as pd
import numpy as np
import string
import re
def Remove(duplicate):   #Removing Duplicate Tweets
    final_list = [] 
    for num in duplicate: 
        if num not in final_list: 
            final_list.append(num) 
    return final_list 

def Cleaning_data(file):
    
#Dataframe
    names=["Created at","Tweets"]

    files_df=pd.read_csv(file,sep=",",names=names,error_bad_lines=False,engine="python")#Reading tweets from csv

    Tweets=[]

    CT2=[] #Cleaned Tweets
    for tweet in files_df['Tweets']:
        
        Tweets.append(' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ", str(tweet)).split()))#Spliting the tweet and replacing the symbols mentioned only with the text.

    for tweet in Tweets:
        CT2.append(re.sub(r'https:...+\w?','',tweet))#Relacing text with https with only the text.
        
    final_list = [] 
    for num in CT2: 
        if num not in final_list: 
            final_list.append(num) 
    Cleaned_1=final_list
    Cleaned_Final=[]
    for tweet in Cleaned_1:
        line=re.sub(r'^RT[\s]+', '', tweet) #Replacing "RT"s from the tweets..
        line=re.sub(r'[:]+', '',line)
        Cleaned_Final.append(line)
    return(Cleaned_Final)

K_Cleaned_data=Cleaning_data("Kohls_search.csv")
M_Cleaned_data=Cleaning_data("Macy.csv")
J_Cleaned_data=Cleaning_data("JC.csv")
```

```python
#Determing Polarity
from textblob import TextBlob 

def get_tweet_sentiment(tweet): 
        ''' 
        Function to classify sentiment of passed tweet 
        using textblob's sentiment method 
        '''
        # create TextBlob object of passed tweet text 
        analysis = TextBlob(tweet) 
        # set sentiment 
        if analysis.sentiment.polarity > 0: 
            return 'positive'
        elif analysis.sentiment.polarity == 0: 
            return 'neutral'
        else: 
            return 'negative'
def get_Polarity(Cleaned_data):
    
    Polarity=[]        
    Final_Polarity=[]
    Polarity_df=pd.DataFrame()
    lst_dic=[]
    cols=["Tweet","FP"]

    for tweet in Cleaned_data:
        
        FP=(get_tweet_sentiment(tweet))   #getting sentiment for each tweets
        vals=[tweet,FP]
        lst_dic.append(dict(zip(cols,vals)))
        Final_Polarity.append(FP)

    Polarity_df=Polarity_df.append(lst_dic)#DataFrame consists of only Tweets and Polarity.
    return(Polarity_df)
K_Polarity_df=get_Polarity(K_Cleaned_data)
M_Polarity_df=get_Polarity(M_Cleaned_data)
J_Polarity_df=get_Polarity(J_Cleaned_data)
```

```python
K_Polarity_df.to_csv('Kohls_Polarity_1.csv',sep=",")#Reading the unsampled DataFrame into a csv for further use.
M_Polarity_df.to_csv('Macys_Polarity_1.csv',sep=",")
J_Polarity_df.to_csv('JCP_Polarity_1.csv',sep=",")


K_Polarity_df= K_Polarity_df.sample(n=280)
M_Polarity_df= M_Polarity_df.sample(n=280)
J_Polarity_df=J_Polarity_df.sample(n=280)



K_Polarity_df.to_csv('Kohls_Polarity.csv',sep=',')#Reading the sampled DataFrame into a csv for further use.
M_Polarity_df.to_csv('Macys_Polarity.csv',sep=',')
J_Polarity_df.to_csv('JCP_Polarity.csv',sep=',')
```
